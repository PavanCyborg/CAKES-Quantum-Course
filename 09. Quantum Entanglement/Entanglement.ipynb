# Brief History of Quantum Entanglement:

- **Einstein** in **1905** was credited for **"Photon as Discrete carrier of Quantum of Energy"**.

- Einstien found the **non-locality** nature of Quantum theory most objectionable.

- In **1935**, a paper was published which explains a Paradox called ***EPR Paradox***. **EPR $\Large\rightarrow$ Einstein, Podolsky and Rosen**. The title of paper is ***"Can Qunatum-Mechanical description of Physical reality be considered complete?***.

- In **1935**, **schrodinger** published a paper on **" Consequences of wavefunction collapse"** which became to be also known as **"Schrodinger's Cat"**. He named the central property of the EPR-Paradox as **Entanglement**.

- In **1964**, **John stewart Bell** started to research on *"Whether there are any local hidden variable theory which can be compatible with Quantum Mechanics?"*. He gave inequalities that must be obeyed by any local hidden variable theory and also shown that those inequalities are violated by Quantum-Systems.
- These Inequalities are also called **"Bell's Inequalities"** and theorem associated with these equations regarding hidden variable theories is called ***"Bell's Theorem"***.

- **John F. Clauser**, commented that the Bell's Inequalities are **not experimentally testable**. He started to research, on finding suitable inequalities that can be experimentally testable. This is where **CHSH-Inequality** is born.

- **Alain Aspect**, showed the violation of Bell's Inequalities experimentally in **1981**.

- **Anton Zeilinger** in 1997, became the first person to demonstrate ***Quantum Teleportation***.

- In **2022**, **Anton Zeilinger**, **Alain Aspect** and **John F. Clauser** got awarded with **Noble Prize** for their contributions in **Quantum Mechanics**.

# Generation of Entangled Polarized Photons in Lab:



This image presents an overview of the process for generating entangled polarized photons in a laboratory setting, specifically through a method known as **Spontaneous Parametric Down-Conversion (SPDC)**. Here's a breakdown of the content:

### **1. Spontaneous Parametric Down-Conversion (SPDC):**
   - **Definition**: SPDC is a quantum optical process where a photon from a laser beam (the "pump" photon) interacts with a nonlinear crystal and is split into two lower-energy photons (signal and idler photons). These photons are entangled, meaning their quantum states are linked, even when separated by large distances.
   - **Equations**:
     - **Energy Conservation**: $\omega_p = \omega_s + \omega_i$, where $\omega_p$ is the frequency of the pump photon, and $\omega_s$ and $\omega_i$ are the frequencies of the signal and idler photons, respectively.
     - **Momentum Conservation**: $\vec{k}_p = \vec{k}_s + \vec{k}_i$, where $\vec{k}_p$, $\vec{k}_s$, and $\vec{k}_i$ represent the momentum vectors of the pump, signal, and idler photons, respectively.

### **2. Diagram Explanation**:
   - **Crystal and Photon Path**: The diagram shows the pump photon entering a nonlinear crystal, where it gets split into two photons with different polarizations—one horizontally polarized and the other vertically polarized. These photons are entangled.
   - **Entangled Photons**: The photons exiting the crystal maintain an entangled state, meaning their properties are correlated even after the split.

### **3. Laboratory Setup**:
   - **Laser Source**: A laser with a wavelength of 405 nm is used as the photon source.
   - **Optical Components**: 
     - **Iris 1**: Used to control the beam diameter or block unwanted light.
     - **Half-Wave Plate**: Used to rotate the polarization of the laser light.
     - **Lenses 1 and 2**: Focus the light beam onto the nonlinear crystal.
     - **Type-1 BBO Crystal**: A nonlinear crystal where the SPDC process occurs.
     - **ND Filters**: Neutral Density filters are used to reduce the intensity of the light.
   - **EMCCD Camera**: An Electron-Multiplying Charge-Coupled Device camera is used to detect the entangled photons.
   
### **4. Results Display**:

This setup and process are key in quantum optics and quantum information experiments, where entangled photons are essential for quantum communication, quantum cryptography, and various quantum computing applications.
