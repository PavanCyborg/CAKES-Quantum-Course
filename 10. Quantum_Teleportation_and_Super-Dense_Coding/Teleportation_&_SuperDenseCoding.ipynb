Consider Alice and Bob share a pair of entangled photon each. The state of Photons is a **"Maximally Entangled State"** which is $\frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)$.

- Then both alice and Bob have got spatially displaced.
- Now alice wanted to send some information to bob. let that state be $\alpha|0\rangle + \beta|1\rangle$.
- Now alice entangles the information state with her photon of EPR pair which is initially shared between alice and bob and then she performs measurement in computational basis.



Then Alice broadcasts her results to the bob. Alice's output determines state of the bob's qubit. Bob takes this message and applies suitable pauli-gate for retrieving the output.



# Super-Dense coding:

- It is reverse of Quantum Teleportation.
- In Quantum Teleportation user can transmit one qubit using two classical bits. Whereas, super-dense coding is a process by which user can transmit two classical bits with one Qubit.
