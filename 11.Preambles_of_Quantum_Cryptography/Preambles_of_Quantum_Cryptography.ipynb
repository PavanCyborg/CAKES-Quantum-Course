- In Quantum world, there are certain rules that ensures the key tampering and monitoring will create disturbance in the transmission and hence these activities are detectable most of time.

# No Cloning Theorem:

It states that **"one cannot duplicate or copy an unknown Quantum State"**.

### **Theorem :**
In a complex Hilbert space $ H $, there does not exist a unitary transformation $ U $ such that it can clone an arbitrary quantum state $ |\psi\rangle $. Specifically, there is no unitary transformation $ U: H \otimes H \rightarrow H \otimes H $ such that:

$$
U \left( |\psi\rangle \otimes |s\rangle \right) = |\psi\rangle \otimes |\psi\rangle, \quad \forall |\psi\rangle \in H
$$

Here, $ |s\rangle $ is some initial state, and the goal is to create two copies of $ |\psi\rangle $ from one.

### **Proof (by contradiction):**

1. **Assume** that such a unitary operation $ U $ exists, which can clone any quantum state. This means:

   $$
   U \left( |\psi\rangle \otimes |s\rangle \right) = |\psi\rangle \otimes |\psi\rangle
   $$

   and similarly for another state $ |\phi\rangle $:

   $$
   U \left( |\phi\rangle \otimes |s\rangle \right) = |\phi\rangle \otimes |\phi\rangle
   $$

2. **Apply** the unitary transformation to these states and then calculate the inner product between the two cloned states:

   $$
   \langle\psi|\phi\rangle = \langle\psi|\langle s|s\rangle| \phi\rangle = \langle \psi | \langle s | U^\dagger U | s \rangle | \phi \rangle = \langle \psi | \langle \psi | \phi \rangle | \phi \rangle = \langle \psi | \phi \rangle^2
   $$

   Since $ U $ is unitary, $ U^\dagger U = I $, and the inner product becomes:

   $$
   \boxed{
   \langle \psi | \phi \rangle^2}
   $$

4. **Analyze the result**: The equation $ \langle \psi | \phi \rangle^2 = \langle \psi | \phi \rangle $ can only be true in two cases:

   - $ \langle \psi | \phi \rangle = 0 $: $ |\psi\rangle $ and $ |\phi\rangle $ are orthogonal.
   - $ \langle \psi | \phi \rangle = 1 $: $ |\psi\rangle $ and $ |\phi\rangle $ are identical.

5. **Conclude**: For $ U $ to clone any state, it would have to work for non-orthogonal states (i.e., states that are not identical and not orthogonal). However, the analysis shows that this is impossible unless the states are orthogonal or identical. This means the assumption that such a unitary operation exists is incorrect.

### **Implication:**
The No-Cloning Theorem implies that quantum information cannot be perfectly copied, which is a stark contrast to classical information. This theorem is fundamental to quantum cryptography and ensures the security of quantum communication protocols since eavesdroppers cannot copy quantum states without disturbing them.

**Non-orthogonal states cannot be descriminated with certainity**

# Heisenberg's Uncertainity principle in Quantum Cryptography :



The image illustrates the application of Heisenberg's Uncertainty Principle in quantum cryptography, using a conceptual setup similar to the double-slit experiment.

### Key Components of the Image:

1. **Alice (Sender) and Bob (Receiver)**:
   - **Alice** is typically the sender in quantum communication, and **Bob** is the receiver. In quantum cryptography, Alice sends quantum particles (like electrons or photons) to Bob.

2. **Electron Source**:
   - Electrons are emitted from a source and move towards a barrier with two slits, labeled **S1** and **S2**. These slits represent different paths the electron can take.

3. **Metal Wall with Slits (S1 and S2)**:
   - The **metal wall** contains two slits (S1 and S2). According to quantum mechanics, if we don’t observe which slit the electron passes through, the electron behaves like a wave and passes through both slits simultaneously, leading to an interference pattern on the detection screen.
   - If we try to observe (measure) which slit the electron passes through, we disturb the system, collapsing the wavefunction and eliminating the interference pattern.

4. **Interference Pattern (P1, P2, P12)**:
   - **P1** and **P2** represent the probability distributions for the electron if it passes through slit S1 or S2, respectively.
   - **P12** shows the interference pattern that results when the electron’s path is not observed. This pattern is a signature of quantum behavior, indicating the electron passed through both slits as a wave.

5. **Quantum Signature and Observation**:
   - The interference pattern (P12) is unique to quantum mechanics and is a "quantum signature."
   - When Bob (the observer) receives the particles and checks the interference pattern, he is observing this quantum signature. If the signature is observed as expected (P12), it indicates that no eavesdropping has occurred. If the pattern is disturbed or missing, it could indicate interference by an eavesdropper, as measuring the quantum system would disturb the pattern.

### Heisenberg’s Uncertainty Principle in This Context:

- **Uncertainty Principle**: In quantum mechanics, Heisenberg’s Uncertainty Principle states that certain pairs of physical properties, like position and momentum, cannot both be known to arbitrary precision. The more accurately one property is known, the less accurately the other can be known.
- **Application in Quantum Cryptography**: In the context of this image, if an eavesdropper tries to measure which path the electron took (i.e., which slit it passed through), this act of measurement introduces uncertainty and disturbs the interference pattern. The very act of trying to observe the electron collapses its wavefunction, removing the interference pattern, thus alerting Alice and Bob to the presence of the eavesdropper.

### Conclusion:

This image effectively demonstrates the principle behind quantum cryptography, particularly how the Uncertainty Principle ensures the security of quantum communication. Any attempt to intercept or measure the quantum particles alters their behavior in a detectable way, providing a means to secure communication against eavesdropping.
